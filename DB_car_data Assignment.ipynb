3.Dump all the data into car table 

```python
#Dumping Data
import csv
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = tuple(i[0].split(','))
        query = "INSERT INTO cardataset.car VALUES {}".format(row_data)
        cursor.execute(query)
```
